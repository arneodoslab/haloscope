```python
# local scripts
import computation as comp # all functions to compute the stack thicknesses and boost spectrum
import optimization as opt # all functions to optimize the stack

# numerical computing
import numpy as np
import pandas as pd
from scipy.interpolate import interp1d
from scipy.optimize import fminbound

#stats
import scipy.stats as stats

# utilities
from time import time
from time import sleep
from copy import deepcopy
from IPython.display import clear_output
import scipy.constants as const


# plotting
import matplotlib.pyplot as plt
import seaborn as sns


SMALL_SIZE = 22
MEDIUM_SIZE = 24
BIGGER_SIZE = 26
LINEWIDTH = 4
MARKERSIZE = 5
FIGSIZE = (9.652,9.652)
palette = sns.color_palette("colorblind")

plt.rc('font', size=BIGGER_SIZE)          # controls default text sizes
plt.rc('axes', titlesize=BIGGER_SIZE)     # fontsize of the axes title
plt.rc('axes', labelsize=BIGGER_SIZE)    # fontsize of the x and y labels
plt.rc('xtick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('ytick', labelsize=BIGGER_SIZE)    # fontsize of the tick labels
plt.rc('legend', fontsize=MEDIUM_SIZE)    # legend fontsize
plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
plt.rc('lines', linewidth=LINEWIDTH)  # thickness of lines
plt.rc('lines', markersize=MARKERSIZE)  # size of the marker
```

```python
# importing the thickness list
N1=1.4656
N2=1.9166
thick=2.736840000000001e-07 ##m
chirp=1.071
nLayers=48

measurementErr = 5*1e-9 #m
thicklistMeasured = np.loadtxt("measured-thickness-080821.txt",skiprows=1)[:,0] * 1e-9 # nm -->m
thicklistTheory = np.load("../theoretical-thickness-stack.npy") * 1e-9 # nm -->m

# we start with just adding +- 5 to the thickness list
thicklistWitherr = (thicklistMeasured+measurementErr,thicklistMeasured-measurementErr)

def get_solution(n1,n2,chirp,d1,N,tlist=[],isMirr=False):
    
    w,sol=comp.solution_rightwave(n1,n2,d1,nLayers=N,chirp=chirp,isTList=tlist,mirror=isMirr)
    phase=comp.freq_to_phase(w,n1,d1)
    wave=comp.freq_to_lamda(w)
    
    return wave,sol

def normalize(data,kind="01"):
    '''
    normalized such that min(data)-->0 and max(data)-->1 if 01, else min(data)-->0 and max(data)-->-1 if 10
    '''
    data = np.array(data)
    if (kind=="01"):
        normalized = (data-min(data))/(max(data)-min(data))
    else:
        normalized = (min(data)-data)/(max(data)-min(data))
    return normalized
```

```python
waveTheory,solTheory =  get_solution(N1,N2,chirp,thick,nLayers,tlist=thicklistTheory,isMirr=True)
waveMeas,solMeas =  get_solution(N1,N2,chirp,thick,nLayers,tlist=thicklistMeasured,isMirr=True)
waveMeasM,solMeasM =  get_solution(N1,N2,chirp,thick,nLayers,tlist=thicklistWitherr[0],isMirr=True)
waveMeasP,solMeasP =  get_solution(N1,N2,chirp,thick,nLayers,tlist=thicklistWitherr[1],isMirr=True)
```

# Method 2 : 90% CL

```python
measlist = deepcopy(thicklistMeasured)
errlist = np.loadtxt("measured-thickness-080821.txt",skiprows=1)[:,1] * 1e-9 # nm -->m # get err b/w 1e-9 and 5e-9
```

```python
errfunc = lambda mu,sigma: np.abs(np.random.normal(mu,sigma))

# simulate new lists of thicknesses by gaussian variations
nsims = 10000
thicksims  = np.zeros(shape=(nsims,len(measlist)))

for ridx in range(0,nsims):
    for cidx,d in enumerate(measlist):
        thicksims[ridx][cidx] = errfunc(measlist[cidx],errlist[cidx])
```

```python
energydistlist = np.zeros(shape=(len(thicksims),len(waveMeas))) #each row corresponds to the boost for each thickness simulation

# create a boost spectrum for each of these simulated thicknesses
t1 = time()
for idx,thicklists in enumerate(thicksims):
    print( str(idx+1)+"/"+str(len(thicksims)) )
    clear_output(wait=True)
    __,energydistlist[idx] =  get_solution(N1,N2,chirp,thick,nLayers,tlist=thicklists,isMirr=True)
t2 = time()
telapsed = t2-t1
print(f"Execution took {telapsed/60:2.2f} minutes")
```

```python
#np.save("10000sim-energydist.npy",energydistlist)
energydistlist = np.load("10000sim-energydist.npy") ## each row corresponds to the boost for each thickness simulation

percentilelist85 = np.percentile(energydistlist,85,axis=0)
percentilelist50 = np.percentile(energydistlist,50,axis=0)
percentilelist15 = np.percentile(energydistlist,15,axis=0)
```

```python
# distribution of max value of boost spectrum across these thicknesses
maxboostlist = np.max(energydistlist,axis=1)
maxboostarglist = np.argmax(energydistlist,axis=1)

avgboostlist = np.mean(energydistlist,axis=0)

plt.figure(figsize=FIGSIZE)
plt.hist(maxboostlist,bins=50,histtype='step')
plt.axvline(max(solMeas),linestyle='--',color='black')
plt.axvline(max(percentilelist15),linestyle='--',color='black')
plt.axvline(max(percentilelist85),linestyle='--',color='black')

plt.xlabel("Boost Factor")
plt.show()
```

```python
# distribution of energy value corresponding to max boost factor
mass= const.h*const.c/(waveMeas*1e-9)*6.2*1e18
plt.figure(figsize=FIGSIZE)
plt.hist(mass[maxboostarglist],bins=50,histtype='step')
plt.xticks(mass[::250])
plt.xlabel("mass [eV/c^2]")
plt.show()
```
