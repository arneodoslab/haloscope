```python
def load_dataset(filename):
    data = np.loadtxt(filename,delimiter=',')
    x = data[:,0]
    y = data[:,1]
    sortedIdx = np.argsort(x)
    x = x[sortedIdx]
    y = y[sortedIdx]
    
    return x,y

massSun,kappaSun = load_dataset("reference-exclusion-curves/an13-stellar/solar-dataset.csv")
massxe10,kappaxe10 = load_dataset("reference-exclusion-curves/an15-xenon10/xe10-dataset.csv")
massxe100,kappaxe100 = load_dataset("reference-exclusion-curves/bloch17-xenon/xe100-dataset.csv")
massCDMShvev, kappaCDMShvev = load_dataset("reference-exclusion-curves/agnese20-cdms/cdms-hvev-dataset.csv")
massNano, kappaNano = load_dataset("reference-exclusion-curves/charaev19-nanowire/nanowire-data.csv")
massSensei, kappaSensei = load_dataset("reference-exclusion-curves/barak20-sensei/sensei-dataset.csv")
massDamic, kappaDamic = load_dataset("reference-exclusion-curves/arevalo20-damic/damic-dataset.csv")
massEdel, kappaEdel = load_dataset("reference-exclusion-curves/arnaud20-edelweiss/edelweiss-dataset.csv")
massTokyo, kappaTokyo = load_dataset("reference-exclusion-curves/suzuki15-tokyo/tokyo-dataset.csv")
massFunk, kappaFunk = load_dataset("reference-exclusion-curves/andrianavalomahefa20-funk/funk-dataset.csv")
```

```python
fig, ax = plt.subplots(figsize=FIGSIZE)

uplim = 1e-8
alphaval = 0.04
ax.plot(massxe10,kappaxe10, color="#85a68d",label="Xe10")
ax.fill_between(massxe10, kappaxe10, color="#85a68d", y2=uplim,alpha=alphaval)

# plt.plot(massxe100,kappaxe100, color="brown",label="Xe100")
# plt.fill_between(massxe100, kappaxe100, color='brown', y2=1e-7,alpha=0.2)

# plt.plot(massxe1T,kappaxe1T, color="k",label="Xe1T")
# plt.fill_between(massxe1T, kappaxe1T, color='k', y2=1e-7,alpha=0.2)

ax.plot(massSun,kappaSun,color=sns.color_palette("colorblind")[8],label="Solar")
ax.fill_between(massSun, kappaSun, color=sns.color_palette("colorblind")[8], y2=uplim,alpha=alphaval)

ax.plot(massNano,kappaNano,color=sns.color_palette("colorblind")[2], label="Nanowire")
ax.fill_between(massNano, kappaNano, color=sns.color_palette("colorblind")[2], y2=uplim,alpha=alphaval)

ax.plot(massSensei,kappaSensei,color=sns.color_palette("colorblind")[9], label="SENSEI")
ax.fill_between(massSensei, kappaSensei, color=sns.color_palette("colorblind")[9], y2=uplim,alpha=alphaval)

ax.plot(massDamic,kappaDamic,color=sns.color_palette("colorblind")[7], label="DAMIC")
ax.fill_between(massDamic,kappaDamic, color=sns.color_palette("colorblind")[7], y2=uplim,alpha=alphaval)

ax.plot(massCDMShvev,kappaCDMShvev,color=sns.color_palette("colorblind")[5], label="CDMS-HVEV")
ax.fill_between(massCDMShvev,kappaCDMShvev, color=sns.color_palette("colorblind")[5], y2=uplim,alpha=alphaval)

ax.plot(massEdel,kappaEdel,color=sns.color_palette("colorblind")[6], label="Edelweiss")
ax.fill_between(massEdel,kappaEdel, color=sns.color_palette("colorblind")[6], y2=uplim,alpha=alphaval)

ax.plot(massTokyo, kappaTokyo,color='#9ea832', label="Tokyo")
ax.fill_between(massTokyo, kappaTokyo, color='#9ea832', y2=uplim,alpha=0.1)

ax.plot(massFunk, kappaFunk,color='#e06373', label="FUNK")
ax.fill_between(massFunk, kappaFunk, color='#e06373', y2=uplim,alpha=alphaval)

#plt.plot(massLocalProjectionRoom,kappaLocalProjectionRoom, color=sns.color_palette("colorblind")[4], label="Halsocope 23")
#plt.fill_between(massLocalProjectionRoom,kappaLocalProjectionRoom, color=sns.color_palette("colorblind")[4], y2=1e-7,alpha=0.2,zorder=1)

ax.plot(massLocal,kappaLocal, color=sns.color_palette("colorblind")[4], label="Haloscope")
ax.fill_between(massLocal,kappaLocal, color=sns.color_palette("colorblind")[4], y2=uplim,alpha=0.06)
ax.fill_between(massLocal,y1=kappaLocal15,y2=kappaLocal85, color=sns.color_palette("colorblind")[-3],alpha=0.6)


# plt.plot(massLocalProjection2C,kappaLocalProjection2C, color=sns.color_palette("colorblind")[5], label="Haloscope",alpha=0.5)
# plt.fill_between(massLocalProjection2C,kappaLocalProjection2C, color=sns.color_palette("colorblind")[5], y2=1e-7,alpha=0.2)


ax.set_yscale("log")
ax.set_xscale("log")
ax.set_ylim(1e-14,uplim)
ax.set_xlim(0.8,13)
ax.set_xlabel("$m_x$ [eV/$c^2$]")
ax.set_ylabel(r'$\mathcal{\chi}$')
ax.tick_params(axis='both', which='major',length=6)
ax.tick_params(axis='both', which='minor',length=3)

y_major = ticker.LogLocator(base = 10.0, numticks = 5)
ax.yaxis.set_major_locator(y_major)
y_minor = ticker.LogLocator(base = 10.0, subs = np.arange(1.0, 10.0) * 0.1, numticks = 10)
ax.yaxis.set_minor_locator(y_minor)

# x_major = ticker.LogLocator(base = 10.0, numticks = 5)
# ax.xaxis.set_major_locator(x_major)
# x_minor = ticker.LogLocator(base = 10.0, subs = np.arange(1.0, 10.0) * 0.01, numticks = 10)
# ax.xaxis.set_minor_locator(x_minor)
ax.set_xticks([1,2, 5, 10])
ax.get_xaxis().set_major_formatter(ticker.ScalarFormatter())



#plt.title("Projected Exclusion Curve",fontsize=20)
#plt.axhline(1e-13)
#plt.legend(loc='lower left')
plt.savefig("bigbadexclusion.svg",dpi=300,bbox_inches = 'tight')
plt.savefig("bigbadexclusion.png",dpi=1000,bbox_inches = 'tight')

plt.show()
```
